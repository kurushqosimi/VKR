```python
import os
from ultralytics import YOLO
import json
import matplotlib.pyplot as plt

# Конфигурация
DATASETS_DIR = r"C:\Users\KosimiK\AI_PROJECTS\fisheye_project\datasets"
RESULTS_DIR = r"C:\Users\KosimiK\AI_PROJECTS\fisheye_project\results"
os.makedirs(RESULTS_DIR, exist_ok=True)

# Список датасетов для обучения
datasets = {
    "original": "coco", 
    "10percent": "coco2017_10percent_fisheye",
    "20percent": "coco2017_20percent_fisheye",
    "30percent": "coco2017_30percent_fisheye",
    "40percent": "coco2017_40percent_fisheye",
    "50percent": "coco2017_50percent_fisheye"
}
```

```python
def train_model(dataset_name, dataset_path):
    model = YOLO("yolov8n.pt").to("cpu")
    
    # Обучение
    results = model.train(
        data=os.path.join(dataset_path, "coco.yaml"),
        epochs=20,
        imgsz=640,
        batch=8,
        name=f"yolov8n_{dataset_name}",
        patience=10,
        device="cpu"
    )
    
    # Сохранение модели
    model.save(os.path.join(RESULTS_DIR, f"yolov8n_{dataset_name}.pt"))
    return results

# Обучение всех моделей
for name, path in datasets.items():
    print(f"\n=== Обучение на {name} ===")
    train_model(name, os.path.join(DATASETS_DIR, path))
```
